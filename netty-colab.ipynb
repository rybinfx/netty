```python
# Init

from netty.render import render
import os

# from google.colab import files
os.makedirs("IN", exist_ok=True)
os.makedirs("OUT", exist_ok=True)
```

```python
# Put your images in "IN" folder
# Collect them from "OUT" folder afer render
# RENDEL ALL IN


BACKGROUND = "" # Initial image ("" for noise)
MIN_SIZE = 9*8, 16*8 # Initial size
MAX_SIZE = 1024 # Max size to stop

STEP = 1.5 # Upscale step
ITERS = 200 # Optimization steps

IMAGES = [[f] for f in os.listdir("IN") if f[0] != "."]
for IMG in IMAGES:
    render(IMG, BACKGROUND, MIN_SIZE, MAX_SIZE, STEP, ITERS)
```
