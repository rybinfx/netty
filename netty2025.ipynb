```python
# Init

from netty.render import render
import os

# from google.colab import files
os.makedirs("IN", exist_ok=True)
os.makedirs("OUT", exist_ok=True)
```

```python
#@title Рендер всех {vertical-output: true}

# RENDEL ALL IN

IMAGES = [[f] for f in os.listdir("IN") if f[0] != "."]
BACKGROUND = ""
MIN_SIZE = 9*8, 16*8
MAX_SIZE = 1024

STEP = 1.5
ITERS = 200

for IMG in IMAGES:
    render(IMG, BACKGROUND, MIN_SIZE, MAX_SIZE, STEP, ITERS)
```
